```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
c_df = pd.read_csv("/canada_per_capita_income.csv")
c_df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(c_df.year, c_df['per capita income (US$)'], color='red', marker='+')
```

```python
new_df = c_df.drop('per capita income (US$)', axis='columns')
new_df
```

```python
prediction_model = linear_model.LinearRegression()
prediction_model.fit(new_df, c_df['per capita income (US$)'])
```

```python
prediction_model.predict(pd.DataFrame([[2020]], columns=['year']))
```
